```python
import pandas as pd
```

**1a. Download WDBC data**

```python
#Column labels
cols = ['ID','Diagnosis','R-M','T-M','P-M','A-M','Sm-M','Com-M','Con-M','ConP-M','Sy-M','F-M','R-SD','T-SD','P-SD','A-SD','Sm-SD','Com-SD','Con-SD','ConP-SD','Sy-SD','F-SD','R-ML','T-ML','P-ML','A-ML','Sm-ML','Com-ML','Con-ML','ConP-ML','Sy-ML','F-ML']
#Import data
data = pd.read_csv("../data/wdbc.data",header=None,names=cols)

#Replace Malignant diagnosis with 1, Benign with 0
data['Diagnosis'].replace('M',1, inplace=True)
data['Diagnosis'].replace('B',0, inplace=True)
```

**1b. Choose first 30 malignant cases and first 50 benign cases for test set and rest as training set**

```python
#Get rows where diagnosis is malignant and benign
M = data.iloc[:,1]==1
B = data.iloc[:,1]==0

#Create training data
train_M_X = data[M].iloc[30:,2:]
train_M_Y = data[M].iloc[30:,1]

train_B_X = data[B].iloc[50:,2:]
train_B_Y = data[B].iloc[50:,1]

train_X = pd.concat([train_M_X,train_B_X],sort=False, axis=0)
train_Y = pd.concat([train_M_Y,train_B_Y],sort=False,axis=0)

#Create testing data
test_M_X = data[M].iloc[:30,2:]
test_M_Y = data[M].iloc[:30,1]

test_B_X = data[B].iloc[:50,2:]
test_B_Y = data[B].iloc[:50,1]

test_X = pd.concat([test_M_X,test_B_X],sort=False,axis=0)
test_Y = pd.concat([test_M_Y,test_B_Y],sort=False,axis=0)
```

**1c. Binary Classification using logistic regression**

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
```

```python
#Fit logistic regression model using training data
log_reg = LogisticRegression(C=1e20)
log_reg.fit(train_X,train_Y)
```

**(2e)(Extra Credit, 15 points) Solve the regression problem of estimating time to recurrence (third attribute) using the next 32 attributes.**

Choose K and report MSE:

```python
#Select 20% of data points from each class in training set
from sklearn.model_selection import train_test_split

#Select data from class R and class N
R = data2[data2['Diagnosis']==1].iloc[0:37,:]
N = data2[data2['Diagnosis']==0].iloc[0:130,:] #.2 is 26 rows out of 130

#Train test split for each class (train=k, test=new training set)
rx_train, rx_test, ry_train, ry_test = train_test_split(R.iloc[:,3:], R.iloc[:,1], test_size=0.2, random_state=0)
nx_train, nx_test, ny_train, ny_test = train_test_split(N.iloc[:,3:], N.iloc[:,1], test_size=0.2, random_state=0)

#Concatenate
k_train_Set_X = pd.concat([rx_train,nx_train], axis =0)
k_train_Set_Y = pd.concat([ry_train,ny_train], axis =0)
k_test_Set_X = pd.concat([rx_test,nx_test], axis =0)
k_test_Set_Y = pd.concat([ry_test,ny_test], axis =0)
```

```python
#Choose best k 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

#Caluclate test error and best K
test_error = [] #Array to store the test errors for each K
min_error = 100 #lowest training error rate
min_k = 100 #K that produces the best training error rate
for i in range(1,33,1):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(k_train_Set_X, k_train_Set_Y)
    pred_i = knn.predict(k_test_Set_X)
    mean_error = mean_squared_error(pred_i, k_test_Set_Y)
    test_error.append(mean_error)
    if(mean_error < min_error): #Update the minimum error and best K
        min_k = i
        min_error = mean_error
best_test_K = min_k
best_test_error = min_error
print("Best Test K:", best_test_K)
print("Best Test Error (raw)",best_test_error)
```

```python
#Test MSE on training set
#repeat using SMOTE to balance
#Training data
train = pd.concat([R,N],sort=False, axis=0)
train_X = train.iloc[:,3:]
train_y = train.iloc[:,1]

RT = data2[data2['Diagnosis']==1].iloc[37:,:]
NT = data2[data2['Diagnosis']==0].iloc[130:,:] #.2 is 26 rows out of 130
test = pd.concat([RT,NT],sort=False,axis=0)
test_X = test.iloc[:,3:]
test_Y = test.iloc[:,1]

knn = KNeighborsRegressor(n_neighbors=best_test_K)
knn.fit(k_train_Set_X, k_train_Set_Y)
pred_i = knn.predict(test_X)
mean_error = mean_squared_error(pred_i, test_Y)
print("MSE: ", mean_error)
```
